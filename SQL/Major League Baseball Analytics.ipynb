```python
%%sql
SELECT *
FROM(SELECT yearID,teamID,W/G AS Winning_Perc,
            RANK() OVER (Partition by yearID ORDER BY Winning_Perc desc) AS RANK
            FROM  Teams
WHERE yearID>=2013 AND yearID<=2016) AS hey
WHERE teamID='ARI';
```

```python
%%sql
SELECT yearID,teamID,RANK() OVER (Partition by yearID ORDER BY HR desc) as Hr_Rank,
       ROUND((H+BB+HBP)/(AB+BB+HBP+SF),2) AS onbase_percentage,
       RANK() OVER (Partition by yearID ORDER BY onbase_percentage desc) AS onbase_rank,
       RANK() OVER (Partition by yearID ORDER BY ERA ASC) as era_rank,
       RANK() OVER (Partition by yearID ORDER BY FP desc) as Fp_Rank, 
       LgWin
FROM   Teams
WHERE  yearID>=2000 AND yearID<=2016;
```

```python
%%sql
# only displaying Cham and NonCham
With t1 as 

(

SELECT yearID,
       teamID,
       RANK() OVER (Partition by yearID ORDER BY HR desc) as Hr_Rank,
       ROUND((H+BB+HBP)/(AB+BB+HBP+SF),2) AS onbase_percentage,
       RANK() OVER (Partition by yearID ORDER BY onbase_percentage desc) AS onbase_rank,
       RANK() OVER (Partition by yearID ORDER BY ERA ASC) as era_rank,
       RANK() OVER (Partition by yearID ORDER BY FP desc) as Fp_Rank, LgWin
FROM  Teams
WHERE yearID>=2000 AND yearID<=2016)

SELECT *
FROM(SELECT LgWin,AVG(Hr_Rank) OVER(Partition BY LgWin),AVG(onbase_rank) OVER(Partition BY LgWin),AVG(era_rank) OVER(Partition BY LgWin),AVG(Fp_Rank) OVER(Partition BY LgWin)
FROM t1
WHERE yearID>=2000 AND yearID<=2016) AS HEY
GROUP BY LgWin;
```

```python
%%sql
With t1
as
(
SELECT *,RANK() OVER (Partition by POS,yearID ORDER BY InnOuts desc) as siralama
FROM Fielding
WHERE InnOuts != '' AND yearID>=2013 AND yearID<=2016 and teamID='ARI' ),

t2 as
(

SELECT *
FROM t1
WHERE siralama=1
GROUP BY yearID,POS),
t3 as

(
select t2.playerID,yearID,POS,siralama,DATE(CONCAT(birthYear,'-',birthMonth,'-',birthDay))  AS DateOfBirth,DATE(CONCAT(yearID,'-',01,'-',01))  AS BegOfYear,TIMESTAMPDIFF(YEAR,DATE(CONCAT(birthYear,'-',birthMonth,'-',birthDay)),DATE(CONCAT(yearID,'-',01,'-',01))) as Age
from t2
inner join Master
on Master.playerID=t2.playerID
ORDER BY yearID)

SELECT yearID,AVG(Age)
FROM t3
GROUP BY yearID;
```

```python
%%sql
WITH T1 AS
(
SELECT *,
                COUNT(*) OVER (PARTITION BY playerID) AS Tot_season,
                CASE
                WHEN teamID = 'FLO' THEN 'MIA'
                WHEN teamID != 'FLO' THEN teamID
                END AS clean_id
FROM     Managers
GROUP BY playerID,
         yearID
HAVING G =162
),
T2 AS 
(
SELECT *
FROM  T1
WHERE T1.Tot_season != 1 AND
      T1.yearID >= 2010  AND
                T1.yearID <= 2016
),
T6 AS
(
SELECT *,
    CASE
                WHEN teamID = 'CHA' THEN 'CHW'
                WHEN teamID = 'CHC' THEN 'CHN'
                WHEN teamID = 'FLO' THEN 'MIA'
                WHEN teamID = 'KCA' THEN 'KCR'
                WHEN teamID = 'LAD' THEN 'LAN'
                WHEN teamID = 'NYA' THEN 'NYY'
                WHEN teamID = 'NYM' THEN 'NYN'
                WHEN teamID = 'SDN' THEN 'SDP'
                WHEN teamID = 'SFG' THEN 'SFN'
                WHEN teamID = 'SLN' THEN 'STL'
                WHEN teamID = 'TBA' THEN 'TBR'
                WHEN teamID = 'WAS' THEN 'WSN'
                ELSE teamID
                END AS clean_id2
FROM Salaries
),
T3 AS
(
SELECT  T2.playerID,
        T2.yearID,
        T2.clean_id,
                T2.W,
                T2.G,
                T2.Tot_season,
        SUM(T6.salary) OVER (PARTITION BY T2.clean_id, T2.yearID) AS Tot_sal
FROM    T2
INNER JOIN T6
ON  T2.clean_id = T6.clean_id2 AND
    T2.yearID = T6.yearID
GROUP BY playerID,
         yearID
),
T4 AS
(
SELECT *,
    T3.W/Tot_sal AS bang
FROM T3
),
T5 AS 
(
SELECT *,
       AVG(T4.bang) OVER (PARTITION BY T4.playerID) AS Avg_bang
FROM  T4

 )
SELECT  T5.playerID,
        T5.yearID,
        Avg_bang
FROM  T5
GROUP BY T5.playerID
ORDER BY Avg_bang DESC
LIMIT 10
;
```
